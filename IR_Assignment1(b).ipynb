```python
import os
import nltk
import nltk.corpus
import string
import pandas as pd
import numpy as np
```

```python
def docs(*doc_path):
    doc_list=[]
    for p in doc_path:
        data=open(p)
        x=data.read()

        punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~''' 

        for i in x:
            if i in punctuations:
                x = x.replace(i, "")
        tokens= [token for token in x.split()]
        
        nltk.download('stopwords')
        from nltk.corpus import stopwords
        sw=stopwords.words('english')
        clean_tokens=[]
        for token in tokens:
            
            wd=token.lower()
            if wd not in sw:
                clean_tokens.append(wd)
        doc_list.append(clean_tokens)
    return doc_list
```
